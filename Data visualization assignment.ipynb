```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
X = iris.loc[:, list(iris.columns[:-1])].values
X[:5]
```

```python
pca = PCA(3)
transformedData = pca.fit_transform(X)
x= transformedData[:, 0]
y = transformedData[:, 1]
z = transformedData[:, 2]
```

```python
plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.set_title("Iris Data 3D Plot", fontsize=15)
ax.set_xlabel('setosa')
ax.set_ylabel('versicolor')
ax.set_zlabel('virginica')
ax.scatter(x, y, z, c=z,cmap='gnuplot2');
plt.show()
```
